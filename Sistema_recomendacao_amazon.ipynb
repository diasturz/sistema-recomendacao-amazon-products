```python
df = pd.read_csv('/content/ratings_Electronics (1).csv')
df.head(10)
```

```python
df.rename(columns={'AKM1MP6P0OYPR':'userId','0132793040':'productId','5.0':'rating','1365811200':'date'},inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'],unit='s')
df
```

```python
user_counts = df['userId'].value_counts().reset_index()
user_counts = user_counts.loc[user_counts['count'] > 50]
user_counts
```

```python
df_active = df[df['userId'].isin(user_counts['userId'])]
df_active
```

```python
pivoted = df_active.pivot_table(index='productId',columns='userId',values='rating')
pivoted
```

```python
pivoted.fillna(0,inplace=True)
pivoted
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

products_sparse = csr_matrix(pivoted.values)
```

```python
model = NearestNeighbors(algorithm='brute')
model.fit(products_sparse)
```

```python
# Product B00LGQ6HL8

distance, suggestion = model.kneighbors(pivoted.filter(items=['B00LGQ6HL8'],axis=0).values,n_neighbors=3)
```
